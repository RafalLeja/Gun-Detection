# Upload lokalnego checkpointu na WandB

Ustaw `CKPT_PATH` na ścieżkę do checkpointu z folderu `logs/` i uruchom komórki poniżej.

```python
CKPT_PATH = "logs/gunmen_rfdetr_finetune_XXXXXXXX_XXXXXX/epoch=0-step=262.ckpt"
```

```python
import wandb

WANDB_ENTITY = "mkarapka-uniwroc"
WANDB_PROJECT = "gun-detection"
MODEL_NAME = "TODO"
```

```python
from pathlib import Path


ckpt_path = Path(CKPT_PATH)
if not ckpt_path.exists():
    raise FileNotFoundError(f"Checkpoint nie istnieje: {ckpt_path}")

run = wandb.init(
    entity=WANDB_ENTITY,
    project=WANDB_PROJECT,
    job_type="upload-checkpoint",
)

artifact = wandb.Artifact(
    name=MODEL_NAME,
    type="model",
    description=f"RF-DETR checkpoint: {ckpt_path.name}",
)
artifact.add_file(str(ckpt_path))
run.log_artifact(artifact)

wandb.finish()
print(f"Checkpoint {ckpt_path.name} uploadowany na WandB.")
```

```python
import wandb
run = wandb.init(entity=WANDB_ENTITY, project=WANDB_PROJECT, job_type='download')
artifact = run.use_artifact(f'{MODEL_NAME}:latest')
path = artifact.download(root='artifacts/')
wandb.finish()
print(f'Pobrano do: {path}')
```
